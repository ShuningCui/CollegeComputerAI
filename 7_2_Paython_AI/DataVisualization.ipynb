在notebook内产生视图

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

线图是所有 **Matplotlib** 图中最简单的。 该图主要用于绘制两组数值之间的关系。 通常，线图用于绘制两个因变量之间的增加或减少趋势。 例如24小时内天气如何变化使用折线图，其中x轴包含每小时信息，y轴包含以度为单位的温度信息。 下面的线图显示0到20之间的平方根。

```python
x_vals = np.arange(0,21)
y_vals = np.sqrt(x_vals)

plt.xlabel('X Values') #设置x轴的标题
plt.ylabel('Y Values')  #y轴标题
plt.title('Square Roots')   #图标题
#画图
plt.plot(x_vals, y_vals)
#有颜色的线，以及线的标注
plt.plot(x_vals, y_vals*2, 'r', label = 'Squrt Root')
#标注的位置
plt.legend(loc = 'upper center')
```

可以使用上一节介绍的pandas库读入csv文件中的数据，并将数据绘制成图形。
下面我们使用了著名的鸢尾花（iris）数据集，它共有4个属性列和一个品种类别列：sepal length（萼片长度）、sepal width（萼片宽度）、petal length（花瓣长度）、petal width （花瓣宽度），单位都是厘米。3个品种类别是Setosa、Versicolour、Virginica，样本数量150个，每类50个。

```python
import pandas as pd
iris_data = pd.read_csv(r"..\DataSet\iris_data.csv")
iris_data.head()
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Sepal vs Petal Length')
plt.scatter(iris_data['sepal_length'], iris_data['petal_length'], c='b')
```

下面使用titanic（泰坦尼克）数据集，记录了真实的登上泰坦尼克号的乘客名单，演示一下柱状图和饼图。
首先导入数据集，数据集中有姓名、性别、年龄、船票号等信息。我们按年龄的分布画出柱状图。

```python
titanic_data = pd.read_csv(r"..\DataSet\titanic_dataset.csv")
titanic_data.head()
plt.title('Age Histogram')
plt.hist(titanic_data['Age'])
plt.show()
```

饼状图：

```python
labels = 'IT', 'Marketing', 'Data Science', 'Finance'
values = [500, 156, 300, 510]
explode = (0.05, 0.05, 0.05, 0.05)
plt.pie(values, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True)
plt.show()
```

结合pandas，也可以这样输出图形：(柱状图和盒状图)

```python
sex_mean = titanic_data.groupby("Sex")["Age"].mean()  #计算不同性别的平均年龄
print(sex_mean)
print(type(sex_mean.tolist()))
df = pd.DataFrame({"Gender":["Female", "Male"], "Age":sex_mean.tolist()})
ax = df.plot.bar(x="Gender", y = "Age", figsize=(8,6))
```

```python
ax = titanic_data.plot.box()  #输出所有数值类型数据的盒状图
```
